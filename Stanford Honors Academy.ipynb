Consider
$$
    f(x) = \frac{3}{2}x + b
$$
In the function above, $b$ is a constant. If $ f(6) = 7$, what is the value of $f(-2)$ ?

    * -5
    * -2
    * 1
    * 7

Observe that we have
$$
f(6) = 7
$$

This is an equation with only one unknown, so:

$$
\frac{3}{2}\cdot(6) + b = 7 \\
\frac{3}{2}\cdot(6) + b -7 = 7 - 7 \\
\frac{3}{2}\cdot(6) - 7 + b = 0 \\
\frac{3}{2}\cdot(6) - 7 + b - b = 0 - b \\
\frac{3}{2}\cdot(6) - 7 \cdot (-1) = -b \cdot (-1) \\
7 - \frac{3}{2}\cdot(6) = b \\
-3 = b 
$$

While preparing to run a marathon, Amelia created a
training schedule in which the distance of her longest
run every week **increased by a constant amount.** If
Amelias training schedule requires that her longest
run in week 4 is a distance of 8 miles and her longest
run in week 16 is a distance of 26 miles, which of the
following best describes how the distance Amelia
runs changes between week 4 and week 16 of her
training schedule?

A) Amelia increases the distance of her longest run
by 0.5 miles each week.

B) Amelia increases the distance of her longest run
by 2 miles each week.

C) Amelia increases the distance of her longest run
by 2 miles every 3 weeks.

D) Amelia increases the distance of her longest run
by 1.5 miles each week

The bold statement, suggest that if we model Amelia training, it must be a linear function, so that:

$$
f(4) = 8 = m\cdot4 + b \\
f(16) = 26 = m\cdot16 + b
$$

Observe that $x$ is a variable with the units of weeks, and such, $m$ must have the units of run per mile. So we are interested on computing the value of $m$.

$$
 8 = m\cdot4 + b \\
 26 = m\cdot16 + b
$$
$$
26 - 8 = 16m - 4m + b -b \\
18 = 12 m \\
\frac{3}{2} = m
$$
